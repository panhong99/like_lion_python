프로그램이란?

어떤 작업을 위해 실행할 수 있는 파일

프로세스란 ? 
컴퓨터에서 연속적으로 실행되고 있는 컴퓨터 프로그램 
메모리에 올라와 실행되고 있는 프로그램의 인스턴스(독립적인 개체)

동적인 개념으로는 실행된 프로그램을 의미

기본적으로 기본적으로 1개의 스레드(메인 스레드)를 가지고 있음


프로세스와 스레드의 차이

프로세스간의 접근
각 프로세스는 별도의 주소공간에서 실행되며 , 한 프로세스는 다른
프로세스의 변수나 자료구조의 접근할 수 없음 

한 프로세스가 다른 프로세스의 자원에 접근하려면 프로세스 간의 통신을 사용해야 함
(파이프 , 파일, 소켓 등을 이용한 통신)

스레드란? 

프로세스 내에서 실행되는 흐름의 단위

프로세스의 특정한 수행 경로 
프로세스가 할당받은 자원을 이용하는 실행의 단위

스레드의 특징 

스레는 프로세스 내에서 각각 stack만 따로 할당받고 code , data , heap영역은 
공유 

스레드느 한 프로세스 내에서 동작되는 여러 실행의 흐름으로 , 프로세스내의
주소 공간이나 자원을(힙 공간 등)을 같은 프로세스 내에 스레드끼리 공유
하면서 실행

프로토콜 = 두개의 프로그램간의 약속


tcp/ip 소켓프로그래밍이란?

가장 최근에 발명된 컴퓨터와 컴퓨터간의 지역네트워크(LAN) 혹은 광역네트워크(WAN)에서 
원활할 통신을 가능하도록 하기 위한 통신규약으로 정의할 수 있다.

소켓: 간단하게 말하면 통신을 위한 일종의 통로 , 두 프로그램이 네트워크를 통해
서로 통신을 할 수 있도록 , 양쪽에 생성되는 링크의 단자

두 소켓이 연결되면 서로 다른 프로세스끼리 데이터 전달가능

소켓이 클라이언트가 붙을때마다 클라이언트소켓이 각각 1대1로 생긴다.

네트워크는 서로 다른 기종의 컴퓨터로 구성되어 있기떄문에 
각 네트워크 간에 공통으로 사용할 수 있는 프로토콜이 필요함 

인터넷에서 컴퓨터간의 통신이 가능하도록 표준화 하여 채택한 통신규약이
tcp/ip

네트워크와 네트워크를 연결하는 데 사용하는 프로토콜인 tcp/ip는 전송 제어
프로토콜 (tcp , transmission control , protocol)과 인터넷 프로토콜
(ip , internet protocol)을 의미

인터넷에서 사용하는 응용 프로그램은 대부분 이 tcp/ip 프로토콜을 이용하여 데이터
를 교환
