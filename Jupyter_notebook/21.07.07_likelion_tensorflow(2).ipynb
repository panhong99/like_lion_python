```python

data = np.loadtxt("/Users/panhong/Desktop/coding_study/Likelion_KDT/Jupyter_notebook/Blood_fat.csv" , delimiter = ",")

x1 = [x_row[0] for x_row in data]
x2 = [y_row[1] for y_row in data]
y_data = [i[2] for i in data]

a1 = tf.Variable(tf.random.uniform([1] , 0 , 10 , dtype = tf.float64 , seed = 0))
a2 = tf.Variable(tf.random.uniform([1] , 0 , 10 , dtype = tf.float64 , seed = 0))
b = tf.Variable(tf.random.uniform([1] , 0 , 100 , dtype = tf.float64 , seed = 0))

def hyper(a1 , a2 , b):
    return a1 * x1 + a2 * x2 + b
    
def costFunc():
    return tf.sqrt(tf.reduce_mean(tf.square(hyper(a1 , a2 , b) - y_data)))

def cost(a1 , a2 , b):
    return tf.sqrt(tf.reduce_mean(tf.square(hyper(a1 , a2 , b) - y_data)))

opt = tf.keras.optimizers.SGD(learning_rate = 0.01)

for i in range(6000):
    opt.minimize(costFunc , var_list = [a1 , a2 , b])
    if i % 100 == 0:
        print(i , f"{cost(a1 , a2 , b)} , {a1.numpy()} , {a2.numpy()} , {b.numpy()}")
print()
da1 = a1.numpy()
da2 = a2.numpy()
db = b.numpy()

for i in range(25):
    print((da1 * x1[i]) + (da2 * x2[i]) + db)
    
```

```python
import tensorflow as tf 
import numpy as np

data = [[2,0] , [4,0] , [6,0]  ,[8,1] , [10,1] , [12,1] , [14,1]]

x_data = [x_row[0] for x_row in data]
y_data = [y_row[1] for y_row in data]

a = tf.Variable(tf.random.normal([1] ,  dtype = tf.float64 , seed = 0))
b = tf.Variable(tf.random.normal([1] ,  dtype = tf.float64 , seed = 0))

def hyoper(a , b):
    return 1/(1 + np.e **  - (a * x_data + b))

# 최솟값 찾기 구문 
def costFunc():
    return - tf.reduce_mean(np.array(y_data) * tf.math.log(hyoper(a , b)) + (1 - np.array(y_data)) * tf.math.log(1 - hyoper(a , b)))

# 코드는 같아보이지만 a, b값을 지속적으로 업데이트해주는 함수
def cost(a , b):
    return - tf.reduce_mean(np.array(y_data) * tf.math.log(hyoper(a , b)) + (1 - np.array(y_data)) * tf.math.log(1 - hyoper(a , b)))
    
opt = tf.keras.optimizers.SGD(learning_rate = 0.5)

for i in range(60001):
    opt.minimize(costFunc , var_list = [a , b])
    if i % 10000 == 0:
        print(i , f"{cost(a,b)} , {a.numpy()} , {b.numpy()}")

new_x_data = 5
y_test = 1 / (1 + np.e ** - (a * new_x_data + b))
print(y_test.numpy())
print("%.40f"% float(y_test.numpy()))

new_x_data = 7
y_test = 1 / (1 + np.e ** - (a * new_x_data + b))
print(y_test.numpy())
print("%.40f"% float(y_test.numpy()))

new_x_data = 13
y_test = 1 / (1 + np.e ** - (a * new_x_data + b))
print(y_test.numpy())
print("%.40f"% float(y_test.numpy()))
```
