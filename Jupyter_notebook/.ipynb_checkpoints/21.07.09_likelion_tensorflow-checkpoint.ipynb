```python
# 노드를 늘려서 계산하는 식
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import numpy as np

x_data = np.array([[0,0] , [0,1] , [1,0] , [1,1]] , dtype = np.float32)
y_data = np.array([[0] , [1] , [1] , [0]] , dtype = np.float32)

# 노드 2개 생성 
x = tf.placeholder(tf.float32 , shape = [None,2] , name = "x-input")
y = tf.placeholder(tf.float32 , shape = [None,1] , name = "y-input")

# 2개가 5개한테 정보 전송
w1 = tf.Variable(tf.random.normal([2 , 5] , name = "weight1"))
b1 = tf.Variable(tf.random.normal([5]) , name = "biae1")
layer1 = tf.sigmoid(tf.matmul(x , w1 ) + b1)

# 5개가 5개한테 정보 전송
w2 = tf.Variable(tf.random.normal([5 , 5] , name = "weight2"))
b2 = tf.Variable(tf.random.normal([5]) , name = "biae2")
layer2 = tf.sigmoid(tf.matmul(layer1 , w2 ) + b2)

# 5개가 5개한테 정보 전송
w3 = tf.Variable(tf.random.normal([5 , 5] , name = "weight3"))
b3 = tf.Variable(tf.random.normal([5]) , name = "biae3")
layer3 = tf.sigmoid(tf.matmul(layer2 , w3 ) + b3)

# 5개가 1개한테 정보 전송 
w4 = tf.Variable(tf.random.normal([5 , 1] , name = "weight4"))
b4 = tf.Variable(tf.random.normal([1]) , name = "biae4")

hypothesis = tf.sigmoid(tf.matmul(layer3 , w4 ) + b4)

cost = -tf.reduce_mean(y * tf.log(hypothesis) + (1 - y) * tf.log(1 - hypothesis))
train = tf.train.GradientDescentOptimizer(learning_rate = 0.01).minimize(cost)

# 소수점으로 된 내용을 1 , 0으로 바꿔주는 함수
predicted = tf.cast(hypothesis > 0.5 , dtype = tf.float32)
# 양쪽을 비교해서 정확도를 측정해주는 함수 
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, y) , dtype = tf.float32))

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(20001):
        sess.run(train ,feed_dict = {x : x_data , y : y_data})
        if step % 5000 == 0: 
            print("step = " , step ,
                 "cost = " , sess.run(cost , feed_dict = {x : x_data , y : y_data}),
                 "w = " , sess.run(w1) , "b = " , sess.run(b1))
    h , c , a = sess.run([hypothesis , predicted , accuracy],
                        feed_dict = {x : x_data , y : y_data})
    print("\nHypothesis : " , h , "\nCorrect : " , c , "\nAccuracy : " , a)
    
        
```
