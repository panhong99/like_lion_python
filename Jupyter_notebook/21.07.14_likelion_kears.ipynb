```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Activation
from tensorflow.keras import optimizers
```

```python
# 피마 인디안 문제(로지스틱 회귀)
seed = 0
# seed로 값 박제
np.random.seed(seed)
tf.random.set_seed(seed)

# 데이터 셋팅
data = np.loadtxt("/Users/panhong/Desktop/coding_study/Likelion_KDT/Jupyter_notebook/csv_file/pima-indians-diabetes.csv" , delimiter = ",")

# 데이터 분류 fit시킬 데이터
training_data = data[:700 , 0:8]
target_data = data[:700 , 8]

# evaluate 시킬 데이터
tr_sample = data[700: , 0:8]
tr_sample_2 = data[700: , 8]

# 입력층 8개 , 노드 30개(렐루) , 노드30개(렐루) , 출력층 1개(시그모이드)
model = Sequential()
model.add(Dense(30 , input_dim = 8 , activation = "relu"))
model.add(Dense(30 , activation = "relu"))
# true , false를 판단할 수 있는 sigmoid함수 
model.add(Dense(1 , activation = "sigmoid"))

# binary가 0,1 구분할때 사용 , adam사용 , metrics = accuracy감시
model.compile(loss = "binary_crossentropy" , optimizer = "adam" , metrics = ["binary_accuracy"])

# 데이터 학습 
model.fit(training_data , target_data , epochs = 1500 , verbose = 1)

# print("\nAccuracy : " , model.evaluate(tr_sample , tr_sample_2)[1])
scores = model.evaluate(tr_sample , tr_sample_2)
print("%s : %.2f%%" % (model.metrics_names[1] , scores[1] * 100))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras import optimizers
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
# from sklearn.preprocession import LabelEncoder

seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)

df = pd.read_csv("/Users/panhong/Desktop/coding_study/Likelion_KDT/Jupyter_notebook/csv_file/iris.csv" , names = ["sepal_length" , "sepal_width" , "petal_length" , "petal_width" , "species"])

sns.pairplot(df , hue = "species");
plt.show()

dataset = df.values
X = dataset[: , 0:4].astype(float)
Y_obj = dataset[: , 4]

# 문자열을 숫자로 변환
e = LabelEncoder()
e.fit(Y_obj)
# 이번에는 Y값이 숫자가 아니라 문자열
# 문자열을 바꿔주려면 클래스 이름을 숫자형태로 바꿔주어야 함
Y = e.transform(Y_obj) # <= 
Y_encoded = tf.keras.utils.to_categorical(Y) #<= 이 함수를 적용하면 여러개의 y의 값을 0,1로만 이루어진 형태로 바꿔줌
# 이 기법을 one-hot-encoding 이라고함

model = Sequential()
# 입력값은 4개고 출력이 1개가 아니라 3개라서 출력층의 노드를 3개를 만들어 주어야함
model.add(Dense(30 , input_dim = 4 , activation = "relu"))
model.add(Dense(30 , activation = "relu"))
model.add(Dense(30 , activation = "relu"))
# 최종 출력이 3개중에 하나여야 함으로 출력층이 3개
# sortmax는 총합이 1인 형태로 바꿔주는 함수
# 3개의 값의 총합을 1.0으로 하고 그 값이 변환이 될때 가장 큰 값을 가장 큰 가중치를 두고
# 그 해당 값이 1 이되고 나머지 값은 0이됨
model.add(Dense(3 , activation = "softmax"))

model.compile(loss = "categorical_crossentropy" , optimizer = optimizers.Adam(lr = 0.01) , metrics = ["accuracy"])
model.fit(X , Y_encoded , epochs = 10 , verbose = 1 , batch_size = 10)

print("\nAccuracy : %.4f" %(model.evaluate(X , Y_encoded)[1]))

plt.figure(figsize = (12 , 12))
sns.heatmap(df.corr() , linewidths = 0.2 , vmax = 0.5 , vmin = 0.1 , cmap = "Greens" , linecolor = "purple" , annot = True) 
plt.show()

```

```python
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras import optimizers
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

df = pd.read_csv("/Users/panhong/Desktop/coding_study/Likelion_KDT/Jupyter_notebook/csv_file/mushrooms.csv")
pd.DataFrame(df , columns = ["cap-shape" , "cap-surface" , "cap-color" , "bruises" , "odor" , "gill-attachment" , "gill-spacing" , "gill-size" , "gill-color" , "stalk-shape" , "stalk-root" , "stalk-surface-above-ring" , "stalk-surface-below-ring" , "stalk-color-above-ring" , "stalk-color-below-ring" , "veil-type" , "veil-color" , "ring-number" , "ring-type" , "spore-print-color" , "population" , "habitat"])
data = pd.DataFrame(columns = df.columns , data = LabelEncoder().fit_transform(df.values.flatten()).reshape(df.shape))
 
Dataset = data.values
X = Dataset[: , 1:].astype(float)
dataset = df.values
Y_obj = dataset[: , 0]
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
Y_encoded = tf.keras.utils.to_categorical(Y)
model = Sequential()
model.add(Dense(30 , input_dim = 22 , activation = "relu"))
model.add(Dense(30 , activation = "relu"))
model.add(Dense(2 , activation = "softmax"))

model.compile(loss = "categorical_crossentropy" , optimizer = "adam" , metrics = ["accuracy"])
model.fit(X , Y_encoded , epochs = 10)

print("\nAccuracy : " , model.evaluate(X , Y_encoded)[1])
```
