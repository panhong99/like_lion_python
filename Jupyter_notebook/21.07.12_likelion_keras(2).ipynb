```python
# 1.xx 버전에서 텐서플로로 수행할 작업을 이렇게나 간소하게 줄여서 코드를 작성했다.
# 폐암 환자 결과 작성하기 구문이다.

# 케라스 코드
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy
import tensorflow as tf

seed = 0
numpy.random.seed(seed)
tf.random.set_seed(seed)

#데이터 수집
Data_set = numpy.loadtxt("/Users/panhong/Desktop/coding_study/Likelion_KDT/Jupyter_notebook/ThoraricSurgery.csv" , delimiter = ",")
# X는 환자의 기록 , Y는 수술결과
X = Data_set[: , 0:17]
Y = Data_set[: , [17]]

# 딥러닝 구조를 결정(모델을 설정하고 실행하는 부분) , 층을 설정하는 부분
model = Sequential()
# model.add의 개수에 따라 layer이 형성된다.
# 은닉층
# Dense = 노드 , input_dim = 입력층 , activation = 활성화 함수
model.add(Dense(30 , input_dim = 17 , activation = "sigmoid"))
# 출력층 , Dense = 노드가 1개 <= 출력층 , 활성화 함수는 시그모이드
# model.add 가 여러개가 있다면 항상 첫번째층이 입력층 + 은닉층이고
# 마지막층은 출력층이 된다. 어떤 활성화 함수를 사용해서 값을 표현할것인지도 결정을 해주어여 한다.
model.add(Dense(1 , activation = "sigmoid"))

# 딥러닝 실행
# 모델이 효과적으로 구현될 수 있게 여러가지 환경설정을 해주는 부분
model.compile(loss = "binary_crossentropy" , optimizer = "adam" , metrics = ["accuracy"])

# epoch = 이 문장을 30번 실행하라는 내용
# batch_size = 전체의 샘플을 10개씩 끊어서 집어넣으라는 내용
# batch_size는 변경가능 
# X = 입력될 값 , Y = 비교할 값
model.fit(X , Y , epochs = 30 , batch_size = 10)

# 결과를 출력 
# print("\n Accuracy : %.4f" % (model.evaluate(X , Y)[1]))
print(model.evaluate(X , Y))
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation , Dense
# learning_rate를 조정하는 방법
# 러닝레이트를 조정해줌으로서 더 좋은 결과를 얻을 수 있다.
from tensorflow.keras import optimizers

training_data = np.array([[0,0] , [0,1] , [1,0] , [1,1]] , "float32")
target_data = np.array([[0] , [1] , [1] , [0]] , "float32")

model = Sequential()

model.add(Dense(4 , input_dim = 2 , activation = "relu"))
model.add(Dense(1 , activation = "sigmoid"))

# keras optimizer을 이용해서 러닝레이트를 조절
opt = optimizers.Adam(lr = 0.1)

model.compile(loss = "binary_crossentropy" , optimizer = "adam" ,
             metrics = ["binary_accuracy"])

# verbose 결과가 표시되는 형태의 종류중 하나
model.fit(training_data , target_data , epochs = 1000 , verbose = 2)

print(model.predict(training_data))
# print("\nAccuracy : %.4f" %(model.evaluate(training_data ,target_data)[1]))
print(model.evaluate(training_data , target_data))
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Activation , Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import RMSprop
from mpl_toolkits.mplot3d import Axes3D

Data_set = np.loadtxt("/Users/panhong/Desktop/coding_study/Likelion_KDT/Jupyter_notebook/Blood_fat.csv" , delimiter = ",")

# 3개의 데이터 셋을 각각 xs ,ys , zs로 분리 
xs = np.array(Data_set[: , 0] , dtype = np.float32)
ys = np.array(Data_set[: , 1] , dtype = np.float32)
zs = np.array(Data_set[: , 2] , dtype = np.float32)

# x_data로 입력값과 결괏값 분리 
x_data = np.array(Data_set[: , 0:2], dtype = np.float32)
y_data = np.array(Data_set[: , 2], dtype = np.float32)

# reshape를 해주는 이유는 그대로 작성을 하게되면 차원이 다르기 때문에 
# 오류가 발생한다.
y_data = y_data.reshape((25 , 1))


rmsprop = RMSprop(lr = 0.01)
model = Sequential()
# 노드는 하나 입력값은 2개(나이 , 혈중지방함량)
model.add(Dense(1 , input_shape=(2,)))
# 선형회귀 자료이기때문에 손실함수는 mse(평균제곱오차) , optimizer은 adam도 상관없음
model.compile(loss = "mse" , optimizer = rmsprop)
model.summary()


# hist에 model.fit을 한 내용을 입력시켜주고 
hist = model.fit(x_data , y_data , epochs = 1000)
# history.keys로 불러온다.
print(hist.history.keys())
# predict를 이용해서 학습된결과에 새로운 값을 대입해서 정상적으로 작동이 되는지 확인
print("100kg 40세의 혈중지방함량치 = " , model.predict(np.array([100 ,40]).reshape(1,2)))
print("60kg 25세의 혈중지방함량치 = " , model.predict(np.array([60 ,25]).reshape(1,2)))

# fit을 한 model값에는 무수히 많은 가중치와 바이어스가 있는데 
# 그 값을 각각 W_ , b_ 에다가 할당해준다.
W_ , b_ = model.get_weights()

# 초기값 , 마지막값 , ~만큼 증가하게 만들기 , 2차원에 요소가 50개인 1차원 배열 1개 
# x , y는 그래프상 좌표를 찍을수 있게 즉 1대1 대응이 가능하도록 차원 및 요소의 개수를 맞춤
x = np.linspace(20, 100 , 50).reshape(50,1)
y = np.linspace(10, 70 , 50).reshape(50,1)

# 차원 찍어보기
print("x = " , x.shape)
print("y = " , y.shape)

# 1차원의 요소 2개를 합쳐서 2차원 배열로 만든다.?
# axis == 두개의 요소를 합치는데 어떤 차원을 기준으로 합칠것인가를 지정해주는 함수 
# 지금은 2가지 요소 다 1차원이므로 설정을 하는것은 의미가 없고 x , y를 서로 합치면 2차원 배열이 탄생한다.
X = np.concatenate((x,y) , axis = 1)
# X 차원 찍어보기
print("X =" , X.shape)

# 가져온 가중치와 바이어스값을 행렬연산
Z = np.matmul(X , W_) + b_

# 밑에서부터는 그래프 그리기
fig = plt.figure(figsize = (12,12))
ax = fig.add_subplot(111 , projection = "3d")

ax.scatter(x , y , Z)

ax.scatter(xs , ys , zs)
ax.set_xlabel("Weight")
ax.set_ylabel("Age")
ax.set_zlabel("Blood fat")
ax.view_init(15 , 15)
plt.show()
```
