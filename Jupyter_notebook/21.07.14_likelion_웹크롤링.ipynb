# 대표적인 웹 크롤링 전체 코드

```python
import requests , json
from datetime import datetime
import requests
from bs4 import BeautifulSoup
import copy
import threading

# 새로운 뉴스를 저장하고 비교할 리스트 2개 생성
oldRtData = [None] * 20
newRtData = [None] * 20
# 몇번 업로드 했는지 업데이트
count = 0
# 파일 열기
outFp = open("date_hot_data" , "w" , encoding = "utf - 8")

# 함수 작동
def get_now():
    # strttime을 이용해서 현재시간을 문자열로 전환
    now = datetime.now().strftime("%X%Y%m%d%H%M")
    # 네이트 사이트에서 실시간 검색어 불러오기
    r = requests.get("http://www.nate.com/js/data/jsonLiveKeywordDataV1.js?v=" + now).content
    # 한글 깨지지 않게 처리
    keyword_list = json.loads(r.decode("euc-kr"))
    # 불러온 결과담을 리스트 생성 
    result = []
    # 불러온 텍스트 가지고 와서 하나씩 Result에 입력
    for k in keyword_list:
        result.append(k[1])
    # result를 출력 
    return result    

# 타이머 및 업데이트 함수 생성
def readRTData():
    # global함수를 이용해 전역함수로 이용
    global oldRtData
    global newRtData
    global count
    
    # newRtdata = get_now함수 불러오기
    newRtData = get_now()
    # 한 번 불러오면 count + 1
    count += 1
    
    # 새롭게 불러온 값과 기존의 값이 같다면 동일하고 count += 1
    if newRtData == oldRtData:
        print("동일" , count)
    else:
        print("변경" , count)
        # 다르다면 newRtdata 출력하고
        print(newRtData)
        # now_T로 현재시각 불러오기
        now_T = datetime.now()
        # 파일에다가 현재 시간 , 분 , 초 입력
        outFp.writelines(str(now_T.hour) + "/" + str(now_T.minute) + "/" + str(now_T.second) +  "/" + " : " + " \n ")
        # 이 함수는 순서가 있는 자료형(list, set, tuple, dictionary, string)을 입력으로 받아 인덱스 값을 포함하는 enumerate 객체를 리턴합니다.
        # 위에 설명이 길지만 enumerate함수는 인덱스값을 포함하는 객체를 출력
        for i , w_string in enumerate(newRtData):
            outFp.writelines(str(i + 1) + " . " + w_string + "\n")
    
    # 1차원 리스트 이상의 copy를 할때는 deepcopy메소드를 사용해야 정확하게 복사가 된다.
    # 어제 나는 계속 데이터가 자동 업데이트가 되버렸는데 지금처럼 비교해야 될 부분을 모두 해주고 난 다음에 업로드를 해주면
    # 어제 나와 같은 문제가 생기지 않는다.
    oldRtData = copy.deepcopy(newRtData)
    
    # 60번 돌리고 60번이 다 되면 끝
    if (count < 60):
        timer = threading.Timer(60 , readRTData)
        timer.start()
    else:
        outFp.close()
        print("=============종료===============")
        
# 함수 시작 
readRTData()
```
